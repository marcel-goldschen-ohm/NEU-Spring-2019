# Diabetes data set

```python
# Load diabetes data set (nonscaled values at https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt)
from sklearn import datasets

diabetes = datasets.load_diabetes()

print(diabetes.DESCR)
```

```python
rawX = rawdf.values[:,:10]

scaledX = (rawX - rawX.mean(axis=0))  # subtract mean to center data around zero
scaledX /= np.sqrt((scaledX**2).sum(axis=0))  # scale to l2 norm (sum of sqaures = 1)
# ... another scaling that is often used is to divide by the standard deviation

X[1, :5], scaledX[1, :5]
```
